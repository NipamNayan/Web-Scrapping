```python
def wiki_header():
    page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
    soup = BeautifulSoup(page.content, 'html.parser')
    for i in soup.find_all('h1'):
        print("MAIN HEADER:")
        print(i.text)
    print("\nSUB-HEADERS:")
    for i in soup.find_all('h2', class_ = "mp-h2"):
        print(i.text)
        
    
```

2. A python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of
release) and save it in form of a CSV file. 

```python
def imdb_top100():
    page = requests.get("https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating")
    Movies_name = []
    soup = BeautifulSoup(page.content, 'html.parser')
    for i in soup.find_all('h3'):
        Movies_name.append(i.text.strip())
    Movies_name2 = []
    Movies_name_cleaned = []
    for i in Movies_name[:100]:
        Movies_name2.append(i.replace('\n','')[:-6])
    for i,j in enumerate(Movies_name2):
        if i < 9:
            k = j[2:]
        elif i > 8 and i < 99:
            k = j[3:]
        else:
            k = j[4:]  
        Movies_name_cleaned.append(k)
    Rating = []
    for i in soup.find_all('div', class_ = "inline-block ratings-imdb-rating"):
        Rating.append(i.text.strip())
    Years = []
    for i in Movies_name[:100]:
        j = i[-6:].split('(')[1]
        k = j.replace(')','')
        Years.append(k)   
    imdb_top_100_movies = pd.DataFrame({})
    imdb_top_100_movies["Name"] = Movies_name_cleaned[:100]
    imdb_top_100_movies["IMDB rating"] = Rating[:100]
    imdb_top_100_movies["Year of release"] = Years[:100]
    return imdb_top_100_movies
```

```python
movie_imdb = imdb_top100()
movie_imdb
```

```python
movie_imdb.to_csv("IMDB_top_100.csv", index= None)  # Saving in csv format
```

3. A python program to display IMDB’s Top rated 100 Indian movies’ data (i.e. Name, IMDB rating, Year
of release) and saving it in form of a CSV file. 

```python
def top_100_indian_movies():
    page = requests.get("https://www.imdb.com/india/top-rated-indian-movies/")
    Movies_name = []
    soup = BeautifulSoup(page.content, 'html.parser')
    indian_movie = []
    movie_name = []
    year = []
    rating = []
    for i in soup.find_all('td',class_ = "titleColumn")[:100]:
        indian_movie.append(i.text.split("      ")[2])
    for i in indian_movie:
        movie_name.append(i.split("\n")[0])
    for i in indian_movie:
        year.append(i.split("\n")[1])
    for i in soup.find_all('td',class_ ="ratingColumn imdbRating")[:100]:
        rating.append(i.text.strip())
    imdb_top_100_indian_movies = pd.DataFrame({})
    imdb_top_100_indian_movies["Name"] = movie_name[:100]
    imdb_top_100_indian_movies["IMDB rating"] = rating[:100]
    imdb_top_100_indian_movies["Year of release"] = year[:100]
    return imdb_top_100_indian_movies
```

```python
ind_top_100 = top_100_indian_movies()
```

4. A python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’

```python
def Book_page_review():
    page = requests.get("https://bookpage.com/reviews")
    soup = BeautifulSoup(page.content, 'html.parser')
    Book_list = []
    for i in soup.find_all('div', class_ ='bp-block article-info')[:5]:
        Book_list.append(i.text.strip())
    Book_name = []
    Author_name = []
    Genre = []
    Book_review = []
    Genre_cleaned = []
    for i in Book_list:
        j = i.split("\n\n\n")
        Book_name.append(j[0])
        Author_name.append(j[1])
        Genre.append(j[2])
        Book_review.append(j[3])
    for i in Genre:
        j = i.split('\n / \n')
        k = '/'.join(j)
        Genre_cleaned.append(k)
    Book_Page = pd.DataFrame({})
    Book_Page["Book Name"] = Book_name
    Book_Page["Author Name"] = Author_name
    Book_Page["Genre"] = Genre_cleaned
    Book_Page["Review"] = Book_review 
    return Book_Page
```

5. A python program to scrape cricket rankings from ‘www.icc-cricket.com’. You have to scrape:
i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
ii) Top 10 ODI Batsmen in men along with the records of their team and rating.
iii) Top 10 ODI bowlers along with the records of their team and rating.

```python
# i) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
def ODI_top_10_team(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    Matches = [soup.find("td", class_ = "rankings-block__banner--matches").text]  #scraping No 1 team from Top line
    Points = [soup.find("td", class_ = "rankings-block__banner--points").text] #scraping No 1 team from Top line
    for i,j in enumerate(soup.find_all("td", class_= "table-body__cell u-center-text")[:18]): #scrapping for remaining 9 teams
        if i%2 == 0:
            Matches.append(j.text)  # Matches Played
        else:
            Points.append(j.text)    # Points
    print("Matches ",Matches)
    print("Points :",Points)
    Teams = []
    for i in soup.find_all("span", class_= "u-hide-phablet")[:10]:
        Teams.append(i.text)
    print("Teams :",Teams)
    Ratings = [soup.find("td", class_= "rankings-block__banner--rating u-text-right").text.strip()] #scraping No 1 from Top line
    for i in soup.find_all("td", class_= "table-body__cell u-text-right rating")[:9]:   #scrapping for remaining 9 teams
        Ratings.append(i.text)
    print("Ratings :",Ratings)
    ODI_team_top10 = pd.DataFrame({})
    ODI_team_top10["Team"] = Teams
    ODI_team_top10["Matches"] = Matches
    ODI_team_top10["Points"] = Points
    ODI_team_top10["Ratings"] = Ratings
    return ODI_team_top10
```

```python
# Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
ODI_top_10_team("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")  #putting url of top 10 ODI men team as input
```

```python
#ii) Top 10 ODI Batsmen in men along with the records of their team and rating.
def ODI_player_Top10(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    Player = [soup.find("td",class_="rankings-block__top-player-container").text.strip()]  # No1 player
    for i in soup.find_all("td", class_="table-body__cell rankings-table__name name")[:9]:  # remaining players
        Player.append(i.text.strip())
    Team = [soup.find("div",class_="rankings-block__banner--nationality").text.strip()]    # No1 player
    for i in soup.find_all("td", class_="table-body__cell nationality-logo rankings-table__team")[:9]:  # remaining players
        Team.append(i.text.strip())
    Rating = [soup.find("div",class_="rankings-block__banner--rating").text.strip()]   # No1 player
    for i in soup.find_all("td", class_="table-body__cell rating")[:9]:      # remaining players
        Rating.append(i.text.strip())
    Top10_ODI_player = pd.DataFrame({})
    Top10_ODI_player["Player"] = Player
    Top10_ODI_player["Team"] = Team
    Top10_ODI_player["Rating"] = Rating
    return Top10_ODI_player
```

```python
def top_10_bowler_ODI():
    page = requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling")
    soup = BeautifulSoup(page.content, 'html.parser')
    Bowlers = [soup.find("div", class_ ="rankings-block__banner--name-large").text]  #No1 bowler
    for i in soup.find_all("td",class_="table-body__cell rankings-table__name name")[:9]:  #remaining top 9
        Bowlers.append(i.text.strip())
    Bowlers = [soup.find("div", class_ ="rankings-block__banner--name-large").text]  #No1 bowler
    for i in soup.find_all("td",class_="table-body__cell rankings-table__name name")[:9]:    #remaining top 9
        Bowlers.append(i.text.strip())
    Nation = [soup.find("div", class_ ="rankings-block__banner--nationality").text.strip()]    #No1 bowler
    for i in soup.find_all("td",class_="table-body__cell nationality-logo rankings-table__team")[:9]:   #remaining top 9
        Nation.append(i.text.strip())
    Rating = [soup.find("div", class_ = "rankings-block__banner--rating").text.strip()]    #No1 bowler
    for i in soup.find_all("td",class_="table-body__cell rating")[:9]:    #remaining top 9
        Rating.append(i.text.strip())
    top10_bowlers = pd.DataFrame({})
    top10_bowlers["Bowler_name"] = Bowlers
    top10_bowlers["Team"] = Nation
    top10_bowlers["Ratings"] = Rating
    return top10_bowlers
```

```python
# Function for Weather Forecast from National Weather Service website of USA
def San_Francisco_weather_forecast():    
    def num_there(s):
        return any(i.isdigit() for i in s)
    Period = []
    Short_desc = []
    Description = []
    page = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.777120000000025&lon=-122.41963999999996#.YCD8RugzZPY")
    soup = BeautifulSoup(page.content, 'html.parser')
    for i in soup.find_all("p",class_="period-name"):
        Period.append(i.text)
    for i in soup.find_all("p",class_="short-desc"):
        Short_desc.append(i.text)
    Info = []
    for i in soup.find("div", {"id": "seven-day-forecast-container"}).find_all("p"):
        Info.append(i.text)
    Tempr = []
    for i in Info:
        if num_there(i) == True:
            Tempr.append(i)
    for i in soup.find_all("div",class_="col-sm-10 forecast-text"):
        Description.append(i.text)
    San_Francisco = pd.DataFrame({})
    San_Francisco["Period"] = Period
    San_Francisco["Short description"] = Short_desc
    San_Francisco["Tempr"] = Tempr
    San_Francisco["Description"] = Description[:9]
    return San_Francisco
```
