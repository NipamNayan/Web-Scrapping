1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome("C:\\chromedriver_win32\\chromedriver")
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
table,class="wikitable sortable jquery-tablesorter"
```

```python
data = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']").text.split('\n')[2:]
```

```python
data
```

```python
data = data[:-1]
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December'] 
```

```python
# Extracting Date from Raw Data
Date_list = []
for i in data:
    for itm in months:
        if itm in i:
            date = itm + ' ' + i.split(itm)[-1]
            Date_list.append(date)
            break  
        else:
            pass
```

```python
len(Date_list)
```

```python
Ranking = []
for i in data:
    Ranking.append(i.split(" ")[0])
```

```python
Ranking
```

```python
Name = []
for lines in data:
    start = lines.find('. "') + len('. "')
    end = lines.find('"[')
    substring = lines[start:end]
    Name.append(substring)
```

```python
# Extracting views from the strings in data
Views_billions = []
for i in data:
    for itm in months:
        if itm in i:
            string1st = i.split(itm)[0]
            Views_billions.append(string1st.split(" ")[-2])    
```

```python
Views_billions
```

```python
Artists = []
temp = []
for i in data:
    for itm in months:
        if itm in i:
            string1st = i.split(itm)[0]
            temp.append(string1st)
for items in temp:
    start = items.find('] ') + len('] ')
    Artists.append(items[start:])            
```

```python
Artists_cleaned = []
for i in Artists:
    j = i.split(" ")[:-2]
    Artists_cleaned.append(" ".join(j))
```

```python
Artists_cleaned
```

```python
Most_viewed_Youtube = pd.DataFrame({})
Most_viewed_Youtube["Rank"] = Ranking
Most_viewed_Youtube["Name"] = Name
Most_viewed_Youtube["Artist"] = Artists_cleaned
Most_viewed_Youtube["Upload_date"] = Date_list
Most_viewed_Youtube["Views(billions)"] = Views_billions
```

```python
Most_viewed_Youtube
```

```python
Data = []
for i in soup.find_all('div',class_="dataTables_wrapper"):
    j = i.find_all("td",class_="data")
    for k in j:
        Data.append(k.text)
```

```python
States = []
for i in soup.find_all('div',class_="dataTables_wrapper"):
    j = i.find_all("td",class_="name")
    for k in j:
        States.append(k.text)
```

```python
Data
```

```python
from itertools import islice

def chunk(it, size):
    it = iter(it)
    return iter(lambda: tuple(islice(it, size)), ())
```

```python
data_df = list(chunk(Data, 6))
```

```python
States = States[:33]
```

```python
list_data_df = []
for i in data_df:
    list_data_df.append(list(i))
```

```python
GDP_indian_states  = pd.DataFrame(list_data_df, columns=["GSDP_19-20(at current prices)","GSDP_18-19(at current prices)","Share(18-19)","GDP(billion)","GSDP_19_20(at 2011-12 prices)","GSDP_18-19(at 2011-12 prices)"])
```

```python
GDP_indian_states = GDP_indian_states[:33]
```

```python
GDP_indian_states["Ranking"] = list(range(1,34))
```

```python
GDP_indian_states["States"] = States
```
