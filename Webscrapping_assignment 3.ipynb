```python
def download_google_image(keyword):
    # Opening the Webdriver
    driver = webdriver.Chrome("C:\\chromedriver_win32\\chromedriver")  
    # Giving the Input Link
    driver.get('https://www.google.com/')
    search = driver.find_element_by_name('q')
    search.send_keys(keyword,Keys.ENTER)
    elem = driver.find_element_by_link_text('Images')
    elem.get_attribute('href')
    elem.click()
    # Scrolling the Pages Automatically to load Images 
    value = 0
    for i in range(20):         # increase range to do more scrolling
        driver.execute_script("scrollBy("+ str(value) +",+1000);")
        value += 1000
        time.sleep(3) 

    elem1 = driver.find_element_by_id('islmp')

    # Creating New Folders for keyword searched to download the image 
    try:
        os.mkdir('downloads_'+keyword)
    except FileExistsError:
        pass

    # Finding the links of the Images
    sub = elem1.find_elements_by_tag_name("img")
    count = 0
    for i in sub:
        if count < 100:      # Change count limit to scrap more images
            src = i.get_attribute('src')
            try:
                if src != None:
                    src  = str(src)
                    print(src)
                    count+=1
                    urllib.request.urlretrieve(src, os.path.join('downloads_'+keyword,'image'+str(count)+'.jpg'))
                else:
                    raise TypeError
            except TypeError:
                print('fail')
```

```python
# Remaining words for which 100 image needs to be scrapped and Downloaded
Words_List = ['cars','Machine Learning']
```

```python
for keywords in Words_List:
    download_google_image(keywords)
```

5. Write a program to scrap geospatial coordinates (latitude, longitude) of a city searched on google maps.

```python
def geospatial_coordinates(cityname): 
    driver = webdriver.Chrome("C:\\chromedriver_win32\\chromedriver")
    driver.get("https://www.mapdevelopers.com/geocode_tool.php")
    driver.find_element_by_class_name('form-control').send_keys(cityname)
    time.sleep(2)
    driver.find_element_by_class_name('form-control').send_keys(Keys.ENTER)
    time.sleep(2)
    latitude = driver.find_element_by_id('display_lat').text
    longitude = driver.find_element_by_id('display_lng').text
    print("The geospatial coordinates (latitude, longitude) of "+cityname+" are:")
    print("Latitude :",latitude)
    print("Longitude :",longitude)
```

```python
geospatial_coordinates("Pune")
```

```python
from selenium.webdriver.common.keys import Keys
```

```python
driver = webdriver.Chrome("C:\\chromedriver_win32\\chromedriver")  
driver.get('https://www.digit.in/')
```

```python
driver.find_element_by_xpath("//div[@class='search']").click()
```

```python
driver.find_element_by_id("globalPageSearchText").send_keys("gaming laptop")
```

```python
driver.find_element_by_id("globalPageSearchText").send_keys(Keys.ENTER)
```

```python
driver.find_element_by_id("content_top10").click()
```

```python
for i in driver.find_elements_by_xpath("//div[@class='searchProduct-desc']"):
    if i.text == "Best Gaming Laptops in India":
        i.click()
        break
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
soup.find_all()
```

```python
details = driver.find_element_by_id("summtable").text.split('\n')
```

```python
Product_name = []
Seller = []
Price = []
for i in details[1:]:
    Price.append(i.split(' ')[-1])
    Seller.append(i.split(' ')[-2])
    Product_name.append(i.split(' ')[:-2])
```

```python
Product_name_final = []
for i in Product_name:
    Product_name_final.append(" ".join(i))
```

```python
top10_gaming_laptop = pd.DataFrame({})
top10_gaming_laptop["Product Name"] = Product_name_final
top10_gaming_laptop["Seller"] = Seller
top10_gaming_laptop["Price"] = Price
top10_gaming_laptop
```
